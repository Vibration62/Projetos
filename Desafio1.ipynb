```python
import pandas
import pyautogui
```

```python
#instalar:
#pyautogui -> automações 
#pandas -> dados
import pyperclip
import time

pyautogui.PAUSE = 1 # diminuir a velocidade da execução do código

#Passo 1: Entrar no sistema da empresa (DRIVE);

pyautogui.press("win") # apertar 1 tecla
pyautogui.write("Opera") # escrever um texto
pyautogui.press("enter")

time.sleep(10) # esperar um tempo determinado após uma linha de código  

link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga" # a criação desse link faz com que alterações futuras do site do sistema seja mais prática
pyautogui.write(link)
pyautogui.press("enter")

time.sleep(12)   

#Passo 2: Navegar no sistema e encontrar a base da dados;

pyautogui.doubleClick(x=431, y=518)

time.sleep(3)
```

```python
#Passo 3: Exportar a base de dados (baixar o arquivo);

pyautogui.click(x=714, y=512)
pyautogui.click(x=1796, y=510)
pyautogui.click(x=1546, y=620)

time.sleep(5)
```

```python
#Passo 4: Calcular os indicadores necessários (faturamento e quantidade);
caminho = r"C:\Users\victo\Downloads\Vendas - Dez.xlsx" # o "r" (cru) funciona para ele não ler o contra barra como algum código e não o caminho do arquivo
tabela = pandas.read_excel(caminho) # lendo e armazenando a base de dados
display(tabela) # ver as informações da base de dados. A formatação da tabela fica melhor com o display - para arquivos ipynb

#somar o faturamento de todos os produtos
faturamento = tabela["Valor Final"].sum()
#somar a quantidade de produtos
qtde_produtos = tabela["Quantidade"].sum()
```

```python
#Passo 5: enviar o e-mail

pyautogui.hotkey("ctrl", "t") # criar nova aba no navegador
pyautogui.write("https://mail.google.com/mail/u/0/#inbox") # entrar no email
pyautogui.press("enter")

time.sleep(10)

pyautogui.click(x=117, y=220) # clicar no botão "escrever"
time.sleep(6)

pyautogui.write("victor2013chagas+diretoria@gmail.com") # enviar o email
pyautogui.press("tab")
pyautogui.press("tab") # passar para o assunto
pyperclip.copy("Relatório de Vendas") # conseguir copiar um texto. Muito utilizado para textos com caracteres especiais
pyautogui.hotkey("ctrl", "v") 
pyautogui.press("tab")

#Corpo do email. O f significa que você quer formatar o texto, podendo utilizar variáveis
texto = f""" 
Prezados,

Segue o relatório de vendas hoje.

Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {qtde_produtos:,}

Qualquer dúvida estou à disposição.
Abs,
Victor Chagas
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

pyautogui.click(x=1162, y=1026) # Enviar o email
```
